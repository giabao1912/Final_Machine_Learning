# GRADIENT DESCENT

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return x**2 + 5*np.sin(x)

def df(x):
    return 2*x + 5*np.cos(x)

# Gradient Descent Function with Full History
def gradient_descent_with_full_history(starting_x, learning_rate, num_iterations):
    x = starting_x
    x_history = [x]  # to store the history of x values
    f_history = [f(x)]  # to store the history of f(x) values
    for i in range(num_iterations):
        grad = df(x)
        x = x - learning_rate * grad
        x_history.append(x)
        f_history.append(f(x))
    return x_history, f_history

# Initial parameters
x0 = -5  # Starting point for gradient descent
learning_rate = 0.1  # Learning rate
num_iterations = 11  # Number of iterations for gradient descent

# Run Gradient Descent for the given number of iterations
x_history, f_history = gradient_descent_with_full_history(x0, learning_rate, num_iterations)

# Generate a range of x values for plotting the function
x_values = np.linspace(-6, 6, 400)
y_values = f(x_values)

# Plotting the function and the gradient descent iterations
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label=r'$f(x) = x^2 + 5\sin(x)$')

# Plot the history of x values
for i, (x, f_val) in enumerate(zip(x_history, f_history)):
    plt.scatter(x, f_val, color='red')  # Plot each x value
    plt.text(x, f_val, f'{i}', fontsize=8)  # Annotate each point with the iteration number

# Set plot labels and title
plt.title('Gradient Descent Iterations on f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

# Print the history for clarity in text form
for iteration, (x, f_val) in enumerate(zip(x_history, f_history)):
    print(f"Iteration {iteration}: x = {x:.4f}, f(x) = {f_val:.4f}")
```

# Momentum 

```python
import numpy as np

# The function and its derivative
def f(x):
    return x**2 + 5*np.sin(x)

def df(x):
    return 2*x + 5*np.cos(x)

# Stochastic Gradient Descent with Momentum
def sgd_with_momentum(starting_x, learning_rate, gamma, num_iterations):
    x = starting_x
    v = 0  # Initialize momentum term
    x_history = [x]  # to store the history of computed x values

    for i in range(num_iterations):
        grad = df(x)
        v = gamma * v + grad  # Momentum update
        x = x - learning_rate * v  # Parameter update
        x_history.append(x)

    return x_history

# Parameters
starting_x = 0  # Start at zero
learning_rate = 0.01  # Small learning rate
gamma = 0.9  # High momentum factor
num_iterations = 50  # Number of iterations

# Perform SGD with Momentum
x_history = sgd_with_momentum(starting_x, learning_rate, gamma, num_iterations)

# Print the last 10 x values to see the convergence
x_history[-10:]
```
